```python
# 01_explore_data.ipynb - Cell 1
from pathlib import Path
import random
from collections import Counter
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import os
plt.rcParams['figure.figsize'] = (10,6)

# notebook path is notebooks/, data is in ../data
DATA_DIR = Path("../data")

```

```python
# Cell 2
list(DATA_DIR.iterdir())
```

```python
# Cell 3
source = DATA_DIR / "PlantVillage"
print("Using source folder:", source)
classes = sorted([p.name for p in source.iterdir() if p.is_dir()])
print("Number of classes detected:", len(classes))
classes[:40]

```

```python
# Cell 4
def count_images(folder):
    return sum(1 for _ in folder.glob("*.*") if _.suffix.lower() in [".jpg",".jpeg",".png"])
class_counts = {c: count_images(source / c) for c in classes}
sorted_counts = sorted(class_counts.items(), key=lambda x: x[1], reverse=True)
sorted_counts[:20]
```

```python
# Cell 5
topk = sorted_counts[:20]
names = [t[0] for t in topk]
vals = [t[1] for t in topk]
plt.barh(range(len(names))[::-1], vals[::-1])
plt.yticks(range(len(names)), names[::-1])
plt.xlabel("Number of images")
plt.title("Top 20 classes by image count")
plt.tight_layout()
plt.show()

```

```python
# Cell 6
from IPython.display import display
sample_classes = classes[:4]
cols = 3
rows = len(sample_classes)
fig, axs = plt.subplots(rows, cols, figsize=(cols*3, rows*3))
for i, cls in enumerate(sample_classes):
    imgs = list((source/cls).glob("*.*"))
    random.shuffle(imgs)
    picks = imgs[:cols]
    for j, p in enumerate(picks):
        img = Image.open(p).convert("RGB")
        axs[i, j].imshow(np.array(img))
        axs[i, j].axis('off')
        if j==0:
            axs[i, j].set_ylabel(cls, fontsize=9)
plt.suptitle("Sample images (3 per class, first 4 classes)")
plt.tight_layout()
plt.show()
```

```python
# Cell 7
import pandas as pd
results_dir = Path("../results")
results_dir.mkdir(parents=True, exist_ok=True)
df = pd.DataFrame(sorted_counts, columns=["class","count"])
df.to_csv(results_dir/"class_counts.csv", index=False)
print("Saved class counts to:", results_dir/"class_counts.csv")
df.head()
```
